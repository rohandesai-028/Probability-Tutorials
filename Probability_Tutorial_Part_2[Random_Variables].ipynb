# Binomial Random Variables

```python
n, p = 50, 0.5  # number of trials, probability of success of each trial

#s generates number of successes in n trials
s = np.random.binomial(n, p, 10)
print(s)

#the mean of a binomial Random Variable is np = 10(0.5) = 5
trials = [10,1000,2000,3000,4000,5000,6000,7000,8000,9000,10000]
res = []
for i in trials:
  s = np.random.binomial(n,p,i)
  res.append(np.average(s))


ax = plt.figure(figsize=(8,6))
plt.plot(trials,res, linewidth=3)
plt.xlabel('Number of Trials',fontsize=18)
plt.ylabel('Mean',fontsize=18)
plt.grid()
plt.show()

```

# Poisson Random Variable

```python
l = 1 # Lambda is the rate parameter at number of events occuring per unit time.
num_of_samples = 10000



for l in range(1,12,3):
    s1 = np.random.poisson(l,num_of_samples)
    fig, ax = plt.subplots(figsize=(9,9))
    ax.hist(s1, bins = [j for j in range(30)],density=True)
    plt.title("PMF of Poisson Random variable for different values of lambda")
    plt.xlabel('k',fontsize=18)
    plt.ylabel('Probability of occurence of k events',fontsize=18)
    plt.grid()
    #l = i
    
```

The PMF of poisson random variable is given by:
$P(X = k) = \frac{e^{-\lambda} \lambda^{k}}{k!}$


For example: if λ = 4 and k = 4, we get $P(X=4) = \frac{4^{4} e^{-4}}{4!} = 0.1954$. We can observe the same result in the second plot at k = 4.
